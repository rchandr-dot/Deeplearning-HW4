

Comparison of FID Scores: ACGAN, DCGAN, and WGAN
The Fretchet Inception Distance (FID) is a popular metric for evaluating the quality of images generated by GANs. It measures the distance between the feature distributions of generated images and real images, with lower scores indicating better image quality and diversity.

```python
import pickle
file = open('Results1/wgan_images/fd_wgan.pickle', 'rb')
file1 = open('Results1/acgan_images/fd_acgan.pickle', 'rb')
file2 = open('Results1/dcgan_images/fd_dcgan.pickle', 'rb')
wgan = pickle.load(file)
dcgan = pickle.load(file2)
acgan = pickle.load(file1)
```

```python
import matplotlib.pyplot as plt

# Apply a different style
plt.style.use('seaborn-darkgrid')  # Try 'ggplot', 'fivethirtyeight', 'bmh', etc.

# Example arrays for demonstration; replace these with your actual data
# dcgan = [values]
# wgan = [values]
# acgan = [values]

plt.plot(dcgan, label='DCGAN')
plt.plot(wgan, label='WGAN')
plt.plot(acgan, label='ACGAN')

plt.title('Fretchet Distance for 50 Epochs')
plt.xlabel('Epochs')
plt.ylabel('Fréchet Distance')
plt.legend()
plt.grid(True)  # Optional, as some themes already include a grid
plt.show()
```
